```python
minkeWhale= open("/content/drive/My Drive/BridgeUp/molecular genetics/1a_CSN2_DNA_txt.txt").read()
hippo= open("/content/drive/My Drive/BridgeUp/molecular genetics/1b_CSN2_DNA_txt (1).txt").read()
boar= open("/content/drive/My Drive/BridgeUp/molecular genetics/1c_CSN2_DNA.txt").read()
print(minkeWhale)
print(hippo)
print(boar)
```

```python
#transcribing the sequences into RNA and assign each RNA sequence to a variable.
codonDict = {"A":"U", "T":"A", "G":"C", "C":"G"}
RNAminkeWhale = ""
RNAhippo = ""
RNAboar = ""

for i in minkeWhale:
    RNAminkeWhale += codonDict[i]
    
for i in hippo:
  RNAhippo += codonDict[i]

for i in boar:
  RNAboar += codonDict[i]

print(RNAminkeWhale)
print(RNAhippo)
print(RNAboar)
```

```python
#amino acid in relation to codon 
codonDict = {"UUU":"F","UUC":"F","UUA":"L","UUG":"L", "CUU":"L","CUC":"L", 
             "CUA":"L","CUG":"L","AUU":"I","AUC":"I","AUA":"I","AUG":"M",
             "GUU":"V","GUC":"V","GUA":"V","GUG":"V","UCU":"S","UCC":"S",
             "UCA":"S","UCG":"S","CCU":"P","CCC":"P","CCA":"P","CCG":"P",
             "ACU":"T","ACC":"T","ACA":"T","ACG":"T","GCU":"A","GCC":"A",
             "GCA":"A","GCG":"A","UAU":"Y","UAC":"Y","UAA":"STOP","UAG":"STOP",
             "UGA":"STOP","CAU":"H","CAC":"H","CAA":"Q","CAG":"Q","AAU":"N",
             "AAC":"N","AAA":"K","AAG":"K","GAU":"D","GAC":"D","GAA":"E",
             "GAG":"E","UGU":"C","UGC":"C","UGG":"W","CGU":"R","CGC":"R",
             "CGA":"R","CGG":"R","AGU":"S","AGC":"S","AGA":"R","AGG":"R",
             "GGU":"G","GGC":"G","GGA":"G","GGG":"G"}
```

```python
#Translate RNA to amino acid

#minke Whale
proteinmw = "" #add protein sequences here
codon1 = "" #only has chunks of three at a time

for i in RNAminkeWhale: #goes through RNAminkeWhale list
    if len(codon1) == 3: #Codons are in groups of three. THus, making the length == 3.
        proteinmw += codonDict[codon1] #matches codon(key) with amino acid(value), and appends it to protein str.
        codon1 = "" #moves to the next block of 3
        codon1 += i #takes i into account before moving on to next iteration.  
    else:
        codon1 += i

#Hippo
proteinh = "" 
codon2 = "" 

for i in RNAhippo: 
    if len(codon2) == 3: 
        proteinh += codonDict[codon2] 
        codon2 = "" 
        codon2 += i 
    else:
        codon2 += i

#boar
proteinb = "" 
codon3 = "" 

for i in RNAboar: 
    if len(codon3) == 3: 
        proteinb += codonDict[codon3]
        codon3 = "" 
        codon3 += i   
    else:
        codon3 += i
```

```python
print(proteinmw)
print(proteinh)
print(proteinb)
```

Similarities in Amino Acid

```python
#similarities in animo acid comparing boar and minke Whale
similarities = 0 #stores the number of similarities
for counter, valuemw in enumerate(proteinmw): #goes through each item in whale protein str.
  valueb = proteinb[counter] #goes through each item in boar protein str.
  if valueb == valuemw: #if the values are the same, 
    similarities += 1#then the counter of how the amount of amino acid increments by one

print(similarities)
```

```python
#similarities in animo acid comparing boar and hippo
similarities1 = 0 #stores the number of similarities
for counter, valueh in enumerate(proteinh): 
  valueb = proteinb[counter]
  if valueb == valueh:
    similarities1 += 1

print(similarities1)
```

```python
#similarities in animo acid comparing minke Whale and hippo
similarities2 = 0 #stores the number of similarities
for counter, valueh in enumerate(proteinh): 
  valuemw = proteinmw[counter]
  if valuemw == valueh:
    similarities2 += 1

print(similarities2)
```

Comparing DNA Sequences

```python
# Boar to minke Whale
simDNA = 0 #stores the number of similarities
for counter, valuemw in enumerate(minkeWhale): 
  valueb = boar[counter]
  if valueb == valuemw:
    simDNA += 1

print(simDNA)
```

```python
#boar to hippo
simDNA1 = 0 #stores the number of similarities
for counter, valueh in enumerate(hippo): 
  valueb = boar[counter]
  if valueb == valueh:
    simDNA1 += 1

print(simDNA1)
```

```python
#hippo to whale
simDNA2 = 0 #stores the number of similarities
for counter, valueh in enumerate(hippo): 
  valuemw = minkeWhale[counter]
  if valuemw == valueh:
    simDNA2 += 1

print(simDNA2)
```
